## What is the minimal level of B-modes in the TOD ?

This is useful to know as this will guide us to find reasonable values of the impact of the number of sub-frequencies in order to optimize the multi-frequency parameters.

We run TOD fabrications with B-modes only maps, no noise (neither photon noise not instrument noise) as a function of tensor-to-scalar ratio to see the level signal (RMS) in the TOD for a given r value

```python
racenter = 0.0      # deg
deccenter = -57.0   # deg
params = {'H0':H0,'omegab':Omegab,'omegac':Omegac,'omegak':0,'scalar_index':0.9624,
             'reion__use_optical_depth':True,'reion__optical_depth':0.0925,
             'tensor_ratio':0.05,'WantTensors':True,'scalar_amp':scalar_amp,'DoLensing':True}

rvals = np.logspace(-3,-1,5)
rmstod = []
for ther in rvals:
    params['tensor_ratio']=ther
    T,E,B,X = pycamb.camb(lmaxcamb+1,**params)
    T[0:50]=0
    ell = np.arange(1,lmaxcamb+1)
    fact = (ell*(ell+1))/(2*np.pi)
    spectra = [ell, T/fact*0, E/fact*0, B/fact, X/fact*0]
    nside = 256
    Nbpixels = 12*nside**2
    scene = QubicScene(nside, kind='IQU')
    mapI,mapQ,mapU=hp.synfast(spectra[1:],nside,new=True, pixwin=True)
    cmb=np.array([mapI,mapQ,mapU]).T
    # No dust
    dust=np.array([mapI*0,mapQ*0,mapU*0]).T
    # Random Sampling
    sampling = create_random_pointings([racenter, deccenter], 100,15)
    # Frquency definition
    dnu_nu_all = 0.25
    nu_min=150*(1.-dnu_nu_all/2)
    nu_max=150*(1.+dnu_nu_all/2)
    effective_duration=1e50   # so that noise is negligible -> need also to specify photon_noise=False of course

    Y,obs=sig.TOD(nu_min,nu_max,0.05,cmb,dust,sampling,scene,effective_duration,verbose=True, photon_noise=False) 
    rmstod.append(np.max(np.std(Y,axis=0)))
```

```python
plot(rvals,rmstod)
xscale('log')
yscale('log')
```

So we can set a value of 2x10-21 as the minimum value of B-modes in the TOD. We will take this as a reference for setting the number of sub-frequencies everywhere.

```python
ref_B_TOD = 2e-21
```

## Exploration of the number of Sub-Frequencies for building the TOD

We loop on values of the number of sub-frequencies to build frequencies and look at the difference with one TOD made with a large number of frequencies. We then choose the one that lead to systematics much smaller than ref_B_TOD defined above...

We use maps with T,E and B power spectra and r=0.02 as well as maps with B only to compare

```python
params['tensor_ratio']=0.02
T,E,B,X = pycamb.camb(lmaxcamb+1,**params)
T[0:50]=0
ell = np.arange(1,lmaxcamb+1)
fact = (ell*(ell+1))/(2*np.pi)
spectra = [ell, T/fact, E/fact, B/fact, X/fact]
spectraBonly = [ell, T/fact*0, E/fact*0, B/fact, X/fact*0]
nside = 256
Nbpixels = 12*nside**2
scene = QubicScene(nside, kind='IQU')
mapI,mapQ,mapU=hp.synfast(spectra[1:],nside,new=True, pixwin=True)
mapI_B,mapQ_B,mapU_B=hp.synfast(spectraBonly[1:],nside,new=True, pixwin=True)
cmb=np.array([mapI,mapQ,mapU]).T
cmb_B=np.array([mapI_B,mapQ_B,mapU_B]).T
# No dust
dust=np.array([mapI*0,mapQ*0,mapU*0]).T
# Random Sampling
sampling = create_random_pointings([racenter, deccenter], 100,15)
```

Definition of frequency bands

```python
dnu_nu_all = 0.25
nu_min=150*(1.-dnu_nu_all/2)
nu_max=150*(1.+dnu_nu_all/2)
# Effective duration
effective_duration=1e50
```

Now the loop:

```python
nn=10
all_subdelta_construction=np.logspace(np.log10(0.005),np.log10(0.25),nn)
allY = []
allY_B = []
for i in xrange(nn):
    print(i,all_subdelta_construction[i])
    Y,obs=sig.TOD(nu_min,nu_max,all_subdelta_construction[i],cmb,dust,sampling,scene,effective_duration,verbose=False)
    allY.append(Y)
    Y_B,obs=sig.TOD(nu_min,nu_max,all_subdelta_construction[i],cmb_B,dust,sampling,scene,effective_duration,verbose=False)
    allY_B.append(Y_B)
```

We now compute the residuals for all bolometer, with respect to the next better spectral resolution. We also compute the residuals with respect to the absolute best resolution. This si calculated for both B only and T,E,B maps.

```python
Absresiduals = np.zeros((992,nn))
Absresiduals_B = np.zeros((992,nn))
residuals = np.zeros((992,nn))
residuals_B = np.zeros((992,nn))
for i in xrange(1,nn-1):
    residuals[:,i] = np.sum((allY[i]-allY[i-1])**2,axis=1)**0.5/len(allY[0][481,:])
    residuals_B[:,i] = np.sum((allY_B[i]-allY_B[i-1])**2,axis=1)**0.5/len(allY_B[0][481,:])
    Absresiduals[:,i] = np.sum((allY[i]-allY[0])**2,axis=1)**0.5/len(allY[0][481,:])
    Absresiduals_B[:,i] = np.sum((allY_B[i]-allY_B[0])**2,axis=1)**0.5/len(allY_B[0][481,:])
residuals.shape
```

```python
plot(all_subdelta_construction, Absresiduals[481,:],label='Absolue T,E,B')
plot(all_subdelta_construction, Absresiduals_B[481,:], label='Absolute T Only')
plot(all_subdelta_construction, residuals[481,:],label='Relative T,E,B')
plot(all_subdelta_construction, residuals_B[481,:], label='Relative T Only')
plot(all_subdelta_construction,all_subdelta_construction*0+ref_B_TOD,'r--', label='Ref_B_TOD')
plot(all_subdelta_construction,all_subdelta_construction*0+ref_B_TOD/5,'k--', label='Ref_B_TOD/5')
xscale('log')
yscale('log')
xlim(np.min(all_subdelta_construction), np.max(all_subdelta_construction))
xlabel('Sub $\Delta\nu/nu$ Construction')
ylabel('Residuals')
legend(loc='upper left', fontsize=16)
```
