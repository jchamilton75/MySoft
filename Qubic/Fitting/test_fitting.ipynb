# Fit with Minuit and MCMC

Let's take the example of a parabola:

```python
x = linspace(-5,8,10)
sy = 1.
y = x**2 + -3*x + 2 + np.random.randn(10)*sy
sigma_y = np.zeros(10) + sy
errorbar(x,y,yerr=sigma_y,fmt='ro')
```

### Fitting with Minuit: just the same

```python
### Define fitting Function
def fonction(x, par):
    return par[0] + par[1]*x + par[2]*x**2

guess = [0., 0., 0.]
res = fitting.do_minuit(x,y,sigma_y,guess,functname=fonction, 
                           fixpars = [False, False, False])
print('Bestfit',res[1])
print('Errorbar',res[2])
print('Covariance',res[3])
```

### Fitting with MCMC just a little bit more ot do
First you need to define the fitting function and the data set as above:

```python
### Same fitting function

### Define Data Set
data = fitting.Data(xvals=x, yvals=y, errors=sigma_y, model=fonction)
```

Then run the MCMC:

```python
### Run MCMC
chain = fitting.run_mcmc(data, ['p0', 'p1','p2'], 
                         niter=80000, nburn=20000)
```
